```python
import pandas as pd

# 1. Load the CSV file into a pandas dataframe
df = pd.read_csv(r'C:/Users/mushr/GEOG300S26LM/genz_social_media_usage_1M.csv')

# 2. Group and calculate mean session minutes
country = (
    df.groupby(['country', 'age'])['avg_session_minutes']
    .mean()
    .reset_index()
)

# 3. Reshape into pivot table
genz_pivot = country.pivot(
    index='country',
    columns='age',
    values='avg_session_minutes'
)

# 4. Display result
print(genz_pivot.head())

# 5. Save to CSV (optional)
genz_pivot.to_csv('social_media_usage_LM.csv')
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

#1. Load the CSV file into a pandas dataframe

df = pd.read_csv('C://Users//mushr//GEOG300S26LM//social_media_usage_LM.csv')


#2. Initialize Nominatim API call with a custom user_agent (required by Nominatim policy)
      # The user_agent should be specific to your application

geolocator = Nominatim(user_agent="lmartinek@student.framingham.edu")

#3. Use RateLimiter to respect the geocoding service's terms of service (e.g., Nominatim)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

#4. Create function to convert country name to latitude, longitude coordinates

def get_coordinates(country_name):
    try:
        location = geocode(country_name)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding {country_name}: {e}")
        return None, None

#5. Apply the function to the 'country' column
df['coordinates'] = df['country'].apply(get_coordinates)

#6. Split the resulting tuples into separate 'latitude' and 'longitude' columns
df[['latitude', 'longitude']] = pd.DataFrame(df['coordinates'].tolist(), index=df.index)

#7. Drop the temporary 'coordinates' column
df = df.drop('coordinates', axis=1)

print(df)
```

```python
df.to_csv("Social_Media_Geopy_LM.csv")
```
